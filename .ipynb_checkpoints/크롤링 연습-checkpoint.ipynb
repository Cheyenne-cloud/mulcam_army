```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://naver.com/").text
soup = BeautifulSoup(res, 'lxml')
```

```python
# HTML 소스를 보기 편하게 
print(soup.prettify())
```

```python
# HTML 소스코드에 해당하는 '태그'가 있는 첫번째 요소
result = soup.find('a')
print(result)
print(result.get_text())
```

```python
# HTML 소스코드에 해당하는 '태그'가 있는 모두 찾기
result2 = soup.find_all('a')

# 리스트형식이므로 for 문을 사용해 텍스트 추출
for site in result2:
    print(site.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.alexa.com/topsites/countries/KR'
html_web_ranking = requests.get(url).text
soup_web_ranking = BeautifulSoup(html_web_ranking, 'lxml')

# p 태그 요소 안에서 a 태그의 요소를 찾기
web_ranking = soup_web_ranking.select('p a')
```

```python
web_ranking[1].get_text()
```

```python
website_ranking = [web_ranking_element.get_text() for web_ranking_element in web_ranking[1:]]
```

```python
website_ranking[:4]
```

```python
import pandas as pd
website_ranking_dict = {'website':website_ranking}
df = pd.DataFrame(website_ranking_dict, columns=['website'], index=range(1,len(website_ranking)+1))
df[0:6]
```

```python
# 멜론 차트 크롤링
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; Touch; rv:11.0) like Gecko'}

url = 'https://www.melon.com/chart/index.htm'
html_melon = requests.get(url, headers = header)
print(html_melon)
html_melon_text = html_melon.text
soup = bs(html_melon_text, 'lxml')

chart_list = soup.select('.lst50,.lst100')
ranks = soup.findAll('div',{'class':'rank01'})
artists = soup.findAll('div',{'class':'rank02'})
like_num = soup.select('tr td div.wrap button span.cnt')
print(like_num[0].text)
```

```python
melon_list = []

for i in chart_list:
    temp = []
    temp.append(i.select_one('.rank').text)
    temp.append(i.select_one('.rank01').a.text)
    temp.append(i.select_one('.rank02').a.text)
    melon_list.append(temp)
```

```python
import csv
with open('melon100.csv', 'w',encoding = 'cp949', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['순위', '곡명','가수'])
    writer.writerow(melon_list)
```
